# Congress API

```python
root = 'https://api.congress.gov/v3'
endpoint = '/member'
params = {'api_key': congresskey2, 'limit': 250}

r = requests.get(root + endpoint,
                 headers=headers,
                 params=params)
r
```

```python
r.json()
```

```python
congressdf = pd.json_normalize(r.json(), record_path = ['members'])
```

```python
congressdf
```

```python
mcquiredf = congressdf[congressdf['name'].str.contains('McGuire')].reset_index()
mcquireID = mcquiredf['bioguideId'][0]
```

```python
mcquireID
```

```python
endpoint = f'/member/{mcquireID}/sponsored-legislation'
```

```python
r = requests.get(root + endpoint,
                 headers=headers,
                 params=params)
r
```

```python
pd.json_normalize(r.json(), record_path = ['sponsoredLegislation'])
```
